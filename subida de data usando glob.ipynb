```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random 
import pickle
import glob
from google.colab import drive

drive.mount('/content/gdrive')
directorio = "gdrive/My Drive/tesis/fotos/partes del rostro"

personas = ["erika","ivan"]

data_entrenamiento = []

def crear_entrenamiento_datos():
        for persona in personas:
            
            ruta = os.path.join(directorio,persona)
            class_num = personas.index(persona)
            print(class_num)
            
#khe = glob.glob(f'gdrive/My Drive/tesis/fotos/partes del rostro/erika/nariz*.*')

            for foto in glob.glob(f'{ruta}/nariz*.*'):
                
                try:
                    
                    matriz_foto = cv2.imread(foto, cv2.IMREAD_GRAYSCALE)
                    #matriz_foto = cv2.imread(os.path.join(ruta,foto), cv2.IMREAD_GRAYSCALE)
                    #matriz_foto = cv2.imread(os.path.join(ruta,foto), cv2.IMREAD_GRAYSCALE)
                    matriz_redimensionada = cv2.resize(matriz_foto, (105,96))
                    #khe = np.array(matriz_redimensionada)
                    #print(khe.shape)
                    data_entrenamiento.append([matriz_redimensionada, class_num])
                    #print(foto)
                    #plt.imshow(matriz_redimensionada, cmap='gray')
                    #plt.show()
                    #print(matriz_redimensionada.shape)
                   
                    
                
                except Exception as e:
                    pass
                   
                    


crear_entrenamiento_datos()

random.shuffle(data_entrenamiento)

x = []
y = []


for caracteristicas, etiqueta in data_entrenamiento:
    x.append(caracteristicas)
    y.append(etiqueta)

# esto es opcional, solo para ver los arreglos print(x[0].reshape(-1, 70, 70, 1))

x = np.array(x).reshape(-1, 105, 96, 1)
y = np.array(y)

print(x.shape)
pickle_out = open("gdrive/My Drive/tesis/data en formato pickle/x2 nariz105x96.pickle","wb")
pickle.dump(x, pickle_out)
pickle_out.close()

pickle_out = open("gdrive/My Drive/tesis/data en formato pickle/y2 nariz105x96.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()

```
